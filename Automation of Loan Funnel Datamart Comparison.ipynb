```python
import pandas as pd

# Load the Excel file
file_path = 'loan_application_datamart_data_UAT.xlsx'
# comparison_df = pd.read_excel(file_path, sheet_name='Comparison').head(3)  # Load only the first 20 rows
sas_df = pd.read_excel(file_path, sheet_name='SAS')
rds_df = pd.read_excel(file_path, sheet_name='RDS')

# Add a 'Remark' column to store mismatched 'cw_txn_id'
comparison_df['Remark'] = ''
```

```python
comparison_df = pd.read_excel(file_path, sheet_name='Comparison').head(75)
#comparison_df = pd.read_excel(file_path, sheet_name='Comparison')
# Add columns for Record Count, Match, and Match Rate
comparison_df['New Record Count'] = 0
comparison_df['New Match'] = 0
comparison_df['New Match Rate'] = 0.0
comparison_df['New Remark'] = ''
```

```python
count = 1
from datetime import datetime

# Function to parse dates and return only the date part (YYYY-MM-DD)
def parse_date(value):
    try:
        if isinstance(value, str):  # If it's a string, try parsing it
            parsed_value = pd.to_datetime(value, errors='coerce').date()  # Convert to date
        elif isinstance(value, (pd.Timestamp, datetime)):  # If it's already a date or timestamp
            parsed_value = value.date()  # Extract only the date part
        else:
            return value  # If it's not a date, return it as is
        
        return parsed_value.strftime('%Y-%m-%d') if parsed_value else None  # Return formatted date (YYYY-MM-DD)
    
    except Exception as e:
        return None


# Loop through each row in the first 20 rows of the Comparison tab
for i, row in comparison_df.iterrows():
    sas_column = row['Variable']
    rds_column = row['RDS Variable Name']
    
    # Check if 'RDS Variable Name' is missing
    if pd.isna(rds_column):
        print(f"Row {i + 1}: Missing RDS Variable Name")
        continue
    
    # Compare data from SAS and RDS
    if sas_column in sas_df.columns and rds_column in rds_df.columns:
        print(f"Row {i + 1}: Columns present in both SAS and RDS")
        
        mismatched_ids = []  # To store all mismatched cw_txn_id values
        
        # Extract SAS and RDS column data
        sas_column_data = sas_df[sas_column]
        rds_column_data = rds_df[rds_column]
        
        # Calculate the Record Count: Number of non-null values
        record_count = sas_column_data.notna().sum()
        
        # Calculate the Match: Number of matched values
        match = 0
        for idx in range(len(sas_column_data)):
            sas_value = sas_column_data.iloc[idx]
            rds_value = rds_column_data.iloc[idx]
            
            if isinstance(sas_value, (pd.Timestamp, datetime, str)) and isinstance(rds_value, (pd.Timestamp, datetime, str)):
                sas_value_converted = parse_date(sas_value)
                rds_value_converted = parse_date(rds_value)
                
                if sas_value_converted == rds_value_converted:
                    match += 1
                    continue
                
                else:
                    # Mismatch in date values
                    cw_txn_id = rds_df['cw_txn_id'].iloc[idx]
                    mismatched_ids.append(str(cw_txn_id))
                    continue
            elif (pd.isna(sas_column_data.iloc[idx]) and pd.isna(rds_column_data.iloc[idx])):
                #match += 1
                continue
            elif sas_column_data.iloc[idx] == rds_column_data.iloc[idx]:
                match += 1
            else:
                # Get the specific 'cw_txn_id' for the mismatched row
                cw_txn_id = rds_df['cw_txn_id'].iloc[idx]
                mismatched_ids.append(str(cw_txn_id))  # Append to the list of mismatched ids
        
        # Calculate the Match Rate
        match_rate = (match / record_count) * 100 if record_count > 0 else 0
        
        # Update the comparison_df with calculated values
        comparison_df.at[i, 'New Record Count'] = record_count
        comparison_df.at[i, 'New Match'] = match
        comparison_df.at[i, 'New Match Rate'] = match_rate
        
        # If there are any mismatched IDs, join them and store in the 'Remark' column
        if mismatched_ids:
            comparison_df.at[i, 'Remark'] = ', '.join(mismatched_ids)

    else:
        print(f"Row {i + 1}: One or both columns not found in SAS or RDS")
```
